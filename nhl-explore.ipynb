```python
import pandas as pd
import numpy as np
import requests, json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# teams URL
# pretty good explanation of the NHL API can be found here https://www.kevinsidwar.com/iot/2017/7/1/the-undocumented-nhl-stats-api
base_url = "https://statsapi.web.nhl.com/api/v1/"
```

```python
# What does teams API looks like
teams_url = base_url + 'teams'
teams_response = requests.get(teams_url)
teams_data = json.loads(teams_response.text)
print(teams_data['teams'][0])
#  create a list for division and conference
division = []
conference = []
for i in range(len(teams_data['teams'])):
    div = teams_data['teams'][i]['division']['name']
    conf = teams_data['teams'][i]['conference']['name']
    division.append(div)
    conference.append(conf)
# create a teams data frame
teams_df = pd.DataFrame(teams_data['teams'])
teams_df['division_value']=division
teams_df['conference_value']= conference
teams_df
```

```python
print(teams_df.iloc[teams_df["firstYearOfPlay"].astype(int).argmax()][['name', 'firstYearOfPlay']]) # the youngest NHL team is Veags Golden Knights
print(teams_df.iloc[teams_df["firstYearOfPlay"].astype(int).argmin()][['name', 'firstYearOfPlay']]) # the oldest is Montreal Canadians
# sns.distplot(teams_df["firstYearOfPlay"].astype(int) )
fig, ax = plt.subplots()
sns.distplot(teams_df["firstYearOfPlay"].astype(int), ax=ax, kde=False)
ax.set_xlim(1909, 2016)
plt.show()
```

```python
# Look at  single season teams stats
people_url = base_url + "teams/15"+"/stats?stats=gameLog"
print(people_url)
people_response = requests.get(people_url)
people_data =json.loads(people_response.text)
print(people_data['stats'])
```
